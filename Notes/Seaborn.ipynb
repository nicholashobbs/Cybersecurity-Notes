```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=25) # shows histogram and kde (kernel density estimation)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')# hue is where you pass categorical
```

```python
sns.rugplot(tips['total_bill']) # kde is a sum of the normals of the rugplot
```

## Categorical Plots

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker') # hue is the power of seaborn
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips, kind='bar') # a more general method for some reason
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tc = tips.corr() # one way to get a matrix form
```

```python
sns.heatmap(tc)
```

```python
fp=flights.pivot_table(index='month', columns='year',values='passengers')
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidth=1)
```

```python
sns.clustermap(fp)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
sns.despine(left=True, bottom=True)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```
